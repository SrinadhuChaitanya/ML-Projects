```python
import seaborn as sns
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix,           # Data
    annot=True,            # Show correlation values
    cmap='coolwarm',       # Color palette
    fmt=".2f",             # Format numbers to 2 decimals
    linewidths=0.5,        # Space between cells
    cbar_kws={'label': 'Correlation Coefficient'} # Color bar label
)
plt.title("Correlation Heatmap of Sales & Advertisement Expenditure", fontsize=14)
plt.show()
```

```python
import numpy as np
```

```python
# Step 4: Create correlation matrix
corr_matrix = df.corr()

# Step 5: Optional – mask the upper triangle for clarity
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Step 6: Plot the heatmap
plt.figure(figsize=(8,6))
sns.heatmap(
    corr_matrix, 
    mask=mask,          # comment this line if you want full square
    annot=True,         # show correlation values
    cmap='coolwarm',    # red-blue gradient
    fmt=".2f",          # 2 decimal places
    linewidths=0.5,     
    cbar_kws={'label': 'Correlation Coefficient'}
)

# Step 7: Add title
plt.title("Correlation Heatmap: Sales vs Advertisement Spend", fontsize=14)
plt.show()
```
